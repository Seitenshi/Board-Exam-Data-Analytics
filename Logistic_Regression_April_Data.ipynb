```python
import pandas as pd
import scipy.stats.contingency
```

```python
url = 'https://raw.githubusercontent.com/Seitenshi/Board-Exam-Data-Analytics/main/BOARD_EXAM_DATA%20-%20APRIL%202023.csv'
my_df = pd.read_csv(url)
```

```python
my_df
```

```python
#change the last column from string to int
my_df['Remarks'] = my_df['Remarks'].replace('FAILED',0)
my_df['Remarks'] = my_df['Remarks'].replace('PASSED',1)

# Drop rows with null values
my_df = my_df.dropna()

my_df
```

```python
#Remarks - # of takes correl 1
table = pd.crosstab(my_df["Remarks"],my_df["# OF TAKES CORREL 1"])

#X - # OF TAKES CORREL 1
#y - Remarks
```

```python
result_table = scipy.stats.contingency.chi2_contingency(table)
```

```python
print('statistics: ', result_table[0])
print('p value: ', result_table[1])
print('degrees of freedom: ', result_table[2])
```
